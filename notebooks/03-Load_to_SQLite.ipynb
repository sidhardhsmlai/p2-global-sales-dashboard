```python
import pandas as pd
import sys
import os
sys.path.append(os.path.abspath('../src'))

# import script that we wrote
import db_utils

print("Setup complete. Utilities imported.")



```

```python
# Define where the CSVs are and where the DB should go
processed_path = '../data/processed'
db_path = '../data/superstore.db'

# Load the CLEAN data we made in Task 1.3
df_orders = pd.read_csv(f'{processed_path}/orders_clean.csv')
df_returns = pd.read_csv(f'{processed_path}/returns_clean.csv')
df_people = pd.read_csv(f'{processed_path}/people_clean.csv')

print("Clean CSVs loaded into memory.")
```

```python
# 1. Create the connection wire
conn = db_utils.create_connection(db_path)

# 2. Send data through the wire
if conn:
    db_utils.load_df_to_sql(df_orders, 'Orders', conn)
    db_utils.load_df_to_sql(df_returns, 'Returns', conn)
    db_utils.load_df_to_sql(df_people, 'People', conn)
    
    # 3. Verify it worked (Ask the DB what tables it has)
    print("\n--- Verification: Tables in Database ---")
    query = "SELECT name FROM sqlite_master WHERE type='table';"
    tables = pd.read_sql_query(query, conn)
    display(tables)
    
    # 4. Close the connection (Hang up the phone)
    conn.close()
    
```
